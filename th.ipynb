```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import r2_score
# Assume df is your DataFrame and 'target' is your target variable
df=pd.read_csv('train_final_th.csv')
# Remove rows with NaN in 'CK' column
df = df.dropna(subset=['TH'])
df = df.replace('\xa0',6)
# Replace infinity values with a large number
df = df.replace([np.inf, -np.inf], np.finfo(np.float32).max)

# Now you can proceed with your code
X = df.drop(columns='TH', axis=1)
y = df['TH']

# Array of random_state values from -100 to 100
random_states = np.arange(start=3050, stop=10000,step=70)

# Initialize a list where we'll store the score of each random_state
scores = []
for state in random_states:

# Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=state)

    # Create a DMatrix from the training data
    dtrain = xgb.DMatrix(X_train, label=y_train)

    # Define the parameters for the XGBoost model
    params = {
        'objective': 'reg:linear',
        'booster': 'gbtree',
        'eta': 0.1,
        'max_depth': 3
    }

    # Train the model
    model = xgb.train(params, dtrain, num_boost_round=100)

    # Create a DMatrix from the test data
    dtest = xgb.DMatrix(X_test)

    # Make predictions
    predictions = model.predict(dtest)
    # Calculate the R^2 score of the predictions
    r2 = r2_score(y_test, predictions)
    scores.append(r2)
results = pd.DataFrame({'random_state':random_states, 'score':scores})
results[results['score'] == results['score'].max()]
```
